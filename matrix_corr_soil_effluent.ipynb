```python
# Importar bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Precisão de 2 casas decimais:
pd.set_option('float_format', '{:.2f}'.format)

# 'Origem' está codificada como 0 e 1 (0 = Soil, 1 = Effluent)
df = pd.read_csv('./data_new_graphics/soil_effluent_matrix.csv', sep=",")
df
```

```python
# Calcular a matriz de correlação de Pearson
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
plot = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", linewidths=.6, cmap='crest', vmin=-1, vmax=1)
plot = plt.title('Correlation coefficients Pearson between variables', fontdict={'fontsize':12}, pad=16)
plt.show()
```

As correlações entre a variável "Origem" e os outros componentes (como Potassium, Sodium, etc.) foram calculadas pela matriz de correlação de Pearson. Vamos entender como chegamos a esses resultados:

---------------
    O que é a correlação de Pearson?

A correlação de Pearson mede a força e a direção da relação linear entre duas variáveis numéricas. Seus valores variam de:

    1: Correlação positiva perfeita (as variáveis aumentam juntas).

    -1: Correlação negativa perfeita (quando uma aumenta, a outra diminui).

    0: Nenhuma correlação linear.

    Como o cálculo foi feito?

Quando incluímos a variável "Origem" (codificada como 0 e 1), ela foi tratada como uma variável numérica. A função corr() do pandas então calculou a correlação entre a coluna "Origem" e cada uma das outras colunas numéricas do dataframe (Phosphorus, Potassium, Sodium, etc.). Aqui está o processo:

    Origem vs. Potassium: Correlação de -0.96.

Esse valor significa que, à medida que a "Origem" muda de 0 (solo) para 1 (efluente), a concentração de potássio tende a ser muito maior no efluente. Como é uma correlação fortemente negativa, isso mostra que, onde a origem é solo (0), o potássio é muito menor.

    Origem vs. Sodium: Correlação de -0.55.

Existe uma correlação negativa moderada, o que significa que o sódio também tende a ser mais presente no efluente (origem = 1).

    Origem vs. Magnesium: Correlação de -0.67.

O magnésio, assim como o sódio e o potássio, tem concentrações maiores no efluente do que no solo, mas a relação não é tão forte quanto com o potássio.

    Origem vs. Phosphorus: Correlação de -0.27.

A correlação entre fósforo e origem é fraca, sugerindo que não há uma diferença tão acentuada entre as concentrações de fósforo no solo e no efluente.

    Origem vs. Calcium: Correlação de 0.15.

Neste caso, há uma correlação muito fraca e positiva, o que indica que o cálcio pode estar um pouco mais presente no efluente, mas a relação é bem fraca.

    Resumo:
A função corr() calculou a correlação de Pearson entre a variável "Origem" (0 ou 1) e as outras variáveis numéricas, resultando nos valores que vimos.

As correlações fortes negativas indicam que certas substâncias (como Potassium e Magnesium) têm concentrações mais altas no efluente, enquanto as correlações mais fracas indicam que outras variáveis não são tão afetadas pela origem.
